```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename = 'Spam_Test1.csv'
raw_data = open(filename, 'rt')
Test1 = np.loadtxt(raw_data, delimiter=",")
```

```python
filename = 'Spam_Test2.csv'
raw_data = open(filename, 'rt')
Test2 = np.loadtxt(raw_data, delimiter=",")
```

```python
filename = 'Spam_Test3.csv'
raw_data = open(filename, 'rt')
Test3 = np.loadtxt(raw_data, delimiter=",")
```

```python
filename = 'Spam_Test4.csv'
raw_data = open(filename, 'rt')
Test4 = np.loadtxt(raw_data, delimiter=",")
```

```python
filename = 'Spam_Test5.csv'
raw_data = open(filename, 'rt')
Test5 = np.loadtxt(raw_data, delimiter=",")
```

```python
x = [10,11,12,13,14,15,16,17,18]
plt.figure(figsize=(15,10))
plt.title('Test Result for Spam Data', size='xx-large')
plt.ylabel('Error', size='x-large')
plt.plot(x,Test1, 'r*-', markersize=10, linewidth=2, label='iSVD')
plt.plot(x,Test2, 'bo-', markersize=10, linewidth=2, label='SSD')
plt.plot(x,Test3, 'gv-', markersize=10, linewidth=2, label='FD')
plt.plot(x,Test4, 'y^-', markersize=10, linewidth=2, label='0.2PFD')
plt.plot(x,Test5, 'c,-', markersize=10, linewidth=2, label='0.8PFD')
plt.tick_params(axis='both', which='major', labelsize=14)
plt.legend(loc=(0.75,0.75), scatterpoints=1)
```
